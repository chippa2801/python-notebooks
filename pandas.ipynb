# Making changes to the data:

```python
# adding new columns in a data frame with some of all the values in a row.
df['total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Generation']
df.head(5)
```

# To delete a variable:


```python
#drop(): to delete a variable.
df = df.drop(columns = ['total'])
print(df.head(3))
```

```python
#method 2: adding new variables:
#sum of rows from 4-9 variable.
df['total'] = df.iloc[:,4:10].sum(axis = 1) # axis = 1 gives horizontally & 0 gives it in vertically.
print(df.head(2))
```

```python
#reordering the variables:
#M1:
#df = df['Name','total'] writing it like this is bit hectc
#M2:storing all the variables in a list.
cols = list(df.columns)
df = df[cols[0:4]+[cols[-1]]+cols[4:12]]
print(df.head(2))
```

# FILTERING THE DATA:

```python
#loc() function:
df.loc[df['Type 1'] == 'Grass']
#for multiple filterings:
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == "Poison")]
#for integer data:
newdf = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == "Poison")|(df['Attack'] == 59)]
newdf
#to reset the new indexes removing the old ones:
newdf = newdf.reset_index()#it saves that old index values as a new variable,to deduct this
#print(newdf)
newdf.reset_index(drop=True)
```

# Filtering based on textual patterns:

```python
df.loc[df['Name'].str.contains('Mega')]
#if we want to do the reverse of it.we can use '~' to get the data which dont have 'mega:
df.loc[~df['Name'].str.contains('Mega')]
#learn regex library for filtering on textual patterns.
```

# Conditional Changes:

```python
df.loc[df['Type 1'] == 'Fire','Type 1'] = 'Flamey'
df
#we can chnge the one condition to other values.
df.loc[df['Type 1'] == 'Flamey','Legendary'] = True
df
#to get the original data:
df = pd.read_csv("modifiedpoki.csv")
df
#modifying things with conditions:
#df.loc[df['total'] >= 450,['Generation','Legendary']] = ['sidhu1','sidhu2']
```

# Aggregate statistics:(groupby):

```python
df
#finding (mean,sum,counting )of classess or groups in a  specific variable
df.groupby(['Type 1']).mean()
#we can also do the sorting for this:
df.groupby(['Type 1']).mean().sort_values(['Defense'],ascending = False)
#sum
df.groupby(['Type 1']).sum()
#for multiple variables:
df.groupby(['Type 1','Type 2']).count()
```
